```python
from urllib.request import urlopen
html = urlopen("https://en.wikipedia.org/wiki/Maruti_Suzuki")
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
#soup.a
#l=soup.find_all('a')
import csv
from datetime import datetime

links = list()

for l in soup.find_all('a',{'href':True}):
      if l ['href'].startswith('http'):
             print (l['href'])
             links.append(l['href'])
             links.append(l.string)
             links.append(datetime.now().strftime("%y-%m-%d-%H-%M"))
```

```python
links
```

```python
import csv

with open('index.csv','a') as output:
    writer = csv.writer(output, delimiter=',')
    for val in links:
        writer.writerows([links])
```
